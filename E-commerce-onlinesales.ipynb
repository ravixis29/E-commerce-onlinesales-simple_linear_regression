# Preparing data for our Model

```python
df.columns
```

```python
X= df[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
X.head()
```

```python
Y =df['Yearly Amount Spent']
Y.head()
```

# Imprting libraries

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.3, random_state=101)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

# Train the model

```python
lm.fit(X_train,Y_train)
```

## Checking the intercept
### basically The linear equation used in simple linear regression is of the form
### y=mx+b
### Where:
### y is the dependent variable (target).
### x is the independent variable (feature).
### m is the coefficient (slope) that the regression model learns.
### b is the intercept term.

```python
print(lm.intercept_)
```

## Check the co-efficient

```python
lm.coef_
```

### Creating a dataframe for the Co-efficient

```python
cdf = pd.DataFrame(lm.coef_,index=X.columns,columns=['coeff'])
```

```python
cdf
```

# Predictions

```python
predictions= lm.predict(X_test)
```

```python
pred = pd.DataFrame(predictions, columns=[' Predicted Amount '])
```

```python
pred.head()
```

## Comparing the prediction with Y_test

```python
Y_test_reset = Y_test.reset_index(drop=True)
```

```python
Y_test_reset.head()
```

```python
result = pd.concat([Y_test_reset, pred], axis=1)
print(result)
```

```python
result = pd.concat([Y_test_reset, pred], axis=1)
print(result.head())
```

```python
plt.figure(figsize=(8, 5))  # Adjust the figure size as needed

plt.scatter(Y_test, pred, color='blue', label='Predicted vs. Actual')
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], linestyle='--', color='red', label='Perfect Prediction Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs. Actual Values')
plt.legend()
plt.show()
```

```python
# calculating the metrics 
from sklearn import metrics 

print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
```

## Plotting Residual

```python
sns.distplot((Y_test-predictions),bins=50,color='black');
```
